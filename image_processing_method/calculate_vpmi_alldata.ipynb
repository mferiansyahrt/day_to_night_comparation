```python
'''
rgb_wraw_noisy = './DATA BARU/RGB_PROCESSED_DATA/WITH_RAW_NOISY/'
rgb_wraw_clean = './DATA BARU/RGB_PROCESSED_DATA/WITH_RAW_CLEAN/'
rgb_sraw_noisy = './DATA BARU/RGB_PROCESSED_DATA/SKIP_RAW_NOISY/'
rgb_sraw_clean = './DATA BARU/RGB_PROCESSED_DATA/SKIP_RAW_CLEAN/'
'''

rgb_wraw_noisy = './RGB_PROCESSED_DATA/WITH_RAW_NOISY/'
rgb_wraw_clean = './RGB_PROCESSED_DATA/WITH_RAW_CLEAN/'
rgb_sraw_noisy = './RGB_PROCESSED_DATA/SKIP_RAW_NOISY/'
rgb_sraw_clean = './RGB_PROCESSED_DATA/SKIP_RAW_CLEAN/'
```

```python
def calculate_on_folder(n_data,folder_path,save_path,excel_name):
    
    # create empty excel file
    path_save = save_path + excel_name
    pd.DataFrame().to_excel(path_save)

    def check_image(num_im,folder_path):
        file = folder_path + f'Image_{num_im}/'
        im = os.listdir(file)
        
        im_new = []
        for i in im:
            if '.jpg' in i:
                im_new.append(i)
            elif '.png' in i:
                im_new.append(i)
            elif '.npy' in i:
                im_new.append(i)
        im_new = sorted(im_new)
        
        return im_new
    
    def vpmi_image(num_im,folder_path,image_name):
        vpmi_show_metrics = True
        # RAW PROCESSED DATA
        file = folder_path + f'Image_{num_im}/'
        # Image on specific name
        file_im = file + image_name
        if '.jpg' in image_name:
            im = cv2.imread(file_im)
        elif '.png' in image_name:
            im = cv2.imread(file_im)
        elif '.npy' in image_name:
            im = np.load(file_im).astype(np.float32)[:,:,[2,1,0]]
            
        aie1,abwf1,aclf1,aal1,lnpf1,ag1,vpmi1= vpmi.calculate_vpmi(im, vpmi_show_metrics)
        val1 = [aie1,abwf1,aclf1,aal1,lnpf1,ag1,vpmi1]
        return val1

    number_of_sheets = 10 #should be    
    data_vpmi = np.zeros((number_of_sheets,n_data,7))
    for i in range(n_data):
        
        im_name = check_image(i,folder_path)
        
        for im in im_name:
            
            #print(im)
            if im == im_name[0]:
                val0 = vpmi_image(i,folder_path,im)  
                data_vpmi[0][i] = val0
                print(f'Done Calculating VPMI on Image {im_name[0]} in Folder Image_{i}')
            elif im == im_name[1]:
                val1 = vpmi_image(i,folder_path,im)
                data_vpmi[1][i] = val1  
                print(f'Done Calculating VPMI on Image {im_name[1]} in Folder Image_{i}')
            elif im == im_name[2]:
                val2 = vpmi_image(i,folder_path,im)
                data_vpmi[2][i] = val2
                print(f'Done Calculating VPMI on Image {im_name[2]} in Folder Image_{i}')
            elif im == im_name[3]:
                val3 = vpmi_image(i,folder_path,im)
                data_vpmi[3][i] = val3 
                print(f'Done Calculating VPMI on Image {im_name[3]} in Folder Image_{i}')
            elif im == im_name[4]:
                val4 = vpmi_image(i,folder_path,im)
                data_vpmi[4][i] = val4
                print(f'Done Calculating VPMI on Image {im_name[4]} in Folder Image_{i}')
            elif im == im_name[5]:
                val5 = vpmi_image(i,folder_path,im)
                data_vpmi[5][i] = val5  
                print(f'Done Calculating VPMI on Image {im_name[5]} in Folder Image_{i}')
            elif im == im_name[6]:
                val6 = vpmi_image(i,folder_path,im)
                data_vpmi[6][i] = val6 
                print(f'Done Calculating VPMI on Image {im_name[6]} in Folder Image_{i}')
            elif im == im_name[7]:
                val7 = vpmi_image(i,folder_path,im)
                data_vpmi[7][i] = val7 
                print(f'Done Calculating VPMI on Image {im_name[7]} in Folder Image_{i}')
            elif im == im_name[8]:
                val8 = vpmi_image(i,folder_path,im)
                data_vpmi[8][i] = val8 
                print(f'Done Calculating VPMI on Image {im_name[8]} in Folder Image_{i}')
            elif im == im_name[9]:
                val9 = vpmi_image(i,folder_path,im)
                data_vpmi[9][i] = val9  
                print(f'Done Calculating VPMI on Image {im_name[9]} in Folder Image_{i}')
            
        print('----')
        
    col = ['AIE','ABWF','ACLF','AAL','LNPF','AG','VPMI']  
    np_sheets_name = []
    for i in range(n_data):
        im = check_image(i,folder_path)
        np_sheets_name.append(im)
    np_sheets_name = np.array(np_sheets_name)
    
    df_list = []
    for data in data_vpmi:
        df = pd.DataFrame(data,columns = col,index = [f'Image_{i}' for i in range(n_data)])
        df_list.append(df)
        
    if np.unique(np_sheets_name).shape[0] == number_of_sheets:
        for i,image in zip(range(len(df_list)),np_sheets_name[0]):
            with pd.ExcelWriter(path_save,mode = 'a') as writer:
                if len(image) > 31:
                    df_list[i].to_excel(writer, sheet_name = image[0:31])
                else:
                    df_list[i].to_excel(writer, sheet_name = image)
        print(f'DONE SAVING {excel_name}!')

    else:
        print('Cannot Input the Data to Excel, probably because there is another method applied on the Image')
        print('See the RGB Processing Code on Conditional Contrast Correction')
        
    return df_list
                    
```

```python
folder_path = [rgb_wraw_noisy,rgb_wraw_clean,rgb_sraw_noisy,rgb_sraw_clean]
save_path = '/Users/m_feriansyahrt/Documents/S2/PCL/Proyek/OLAH_DATA_VPMI/'
excel_name = ['RGB_WITH_RAW_NOISY.xlsx','RGB_WITH_RAW_CLEAN.xlsx',
              'RGB_SKIP_RAW_NOISY.xlsx','RGB_SKIP_RAW_CLEAN.xlsx']

for folder,excel in zip(folder_path,excel_name):
    print(folder)
    print(excel)
    calculate_on_folder(n_data,folder,save_path,excel)
    print('***********')
```
