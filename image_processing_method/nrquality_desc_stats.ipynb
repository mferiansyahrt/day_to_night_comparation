```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
nr_lama = './OLAH_DATA_NRQUALITY/'
nr_baru = './OLAH_DATA_NRQUALITY_DATABARU/'

vpmi_lama = './OLAH_DATA_VPMI/'
vpmi_baru = './OLAH_DATA_VPMI_DATABARU/'
```

```python
filename_nr_lama = sorted(os.listdir(nr_lama))
filename_nr_baru = sorted(os.listdir(nr_baru))

filename_vpmi_lama = sorted(os.listdir(vpmi_lama))
filename_vpmi_baru = sorted(os.listdir(vpmi_baru))
```

```python
if '.DS_Store' in filename_nr_lama:
    os.remove(nr_lama + '.DS_Store')
elif '.DS_Store' in filename_nr_baru:
    os.remove(nr_baru + '.DS_Store')
else:
    pass
    
if '.DS_Store' in filename_vpmi_lama:
    os.remove(vpmi_lama + '.DS_Store')
elif '.DS_Store' in filename_vpmi_baru:
    print(True)
    os.remove(vpmi_baru + '.DS_Store')
else:
    pass
```

```python
display(filename_nr_lama)
display(filename_nr_baru)

display(filename_vpmi_lama)
display(filename_vpmi_baru)
```

# NR QUALITY

```python
def desc_stat(path,filename):
    alldata_excel = path + filename
    nr_data = pd.read_excel(alldata_excel)
    nr_data.index = nr_data['Row']
    nr_data = nr_data.drop('Row',axis = 1)
    return nr_data,nr_data.describe()
```

```python
print('NR QUALITY DARI DATA LAMA\n')

for i in range(len(filename_nr_lama)):
    file = filename_nr_lama[i]
    print(file)
    df,df_stat = desc_stat(nr_lama,file)
    display(df)
    display(df_stat)
    print('-'*10)
```

```python
print('NR QUALITY DARI DATA BARU\n')

for i in range(len(filename_nr_baru)):
    file = filename_nr_baru[i]
    print(file)
    df,df_stat = desc_stat(nr_baru,file)
    display(df)
    display(df_stat)
    print('-'*10)
```

# VPMI

```python
def desc_stat_vpmi(path,filename):
    alldata_excel = path + filename
    
    excel_file = pd.ExcelFile(alldata_excel)
    sheets_name = excel_file.sheet_names
    
    if 'Sheet1' in sheets_name:
        sheets_name.remove('Sheet1')
        
    sheets_name = sorted(sheets_name)
    
    for sheet in sheets_name:
        nr_data = pd.read_excel(alldata_excel,index_col = 0,sheet_name = sheet)
        print(f'Excel file on {filename} displayng data on sheet "{sheet}"')
        display(nr_data)
        display(nr_data.describe())
        print('\n')
        
        
```

```python
print('VPMI QUALITY DARI DATA LAMA\n')

for i in range(len(filename_vpmi_lama)):
    file = filename_vpmi_lama[i]
    
    desc_stat_vpmi(vpmi_lama,file)
    
    print('-'*10)
```

```python
print('VPMI QUALITY DARI DATA BARU\n')

for i in range(len(filename_vpmi_baru)):
    file = filename_vpmi_baru[i]
    
    desc_stat_vpmi(vpmi_baru,file)
    
    print('-'*10)
```
